```python
## Problem 1: List the names of top 10 actors who have acted in most films, include the number of films.
```

```python
%%sql 

SELECT * FROM film_actor
```

```python
%%sql

SELECT actor_id, CONCAT(first_name,',' ,last_name) AS actor_name, COUNT(actor_id) AS number_of_films
FROM actor INNER JOIN film_actor USING(actor_id)
GROUP BY actor_id
ORDER BY number_of_films DESC
LIMIT 10
```

```python
## Problem 2: List the titles of all films on the inventory of Store 2 only, not found in store 1
```

```python
%%sql 

SELECT DISTINCT film.title, inventory.store_id FROM inventory 
INNER JOIN film USING(film_id)
WHERE store_id <> 1
```

```python
## Problem 3: List the names of all actors who have co-starred in the same movies with Natalie Hopkins.
```

```python
%%sql 

WITH q AS 
(SELECT actor_id, film_id
FROM actor JOIN film_actor USING(actor_id)
WHERE first_name = 'NATALIE' AND last_name = 'HOPKINS')

SELECT DISTINCT actor_id, CONCAT(first_name, ',', last_name) AS actor_name
FROM actor JOIN film_actor USING(actor_id)
WHERE actor_id NOT IN(SELECT actor_id FROM q)
AND film_id IN (SELECT film_id FROM q)
```

```python
## Problem 6: List all orders by employees on their own birthdays. 
```

```python
%%sql 

SELECT order_number, order_date, employee_birthdate 
FROM employee NATURAL LEFT JOIN `order` 
WHERE MONTH(order_date) = MONTH(employee_birthdate) 
AND DAYOFMONTH(order_date) = DAYOFMONTH(employee_birthdate)
```

```python
## Problem 7: Show the date with highest number of orders
```

```python
## Problem 9: List the names of Products that are offered by only a single vendor, inlcuding the name of that vendor. 
```

```python
%%sql 

WITH q as 
(SELECT product_number, COUNT(vendor_id) as number_of_vendors
FROM product_vendor
GROUP BY product_number
HAVING number_of_vendors = 1)

SELECT product.product_number, product.product_name, vendor.vendor_name
FROM product NATURAL JOIN q 
NATURAL JOIN product_vendor
NATURAL JOIN vendor 
```

```python
## Problem 10: List all products, including the name of the vendor that offers it at the lowest price. 
```

```python
%%sql 

WITH q as(
SELECT MIN(wholesale_price) as lowest_price 
FROM product_vendor
GROUP BY product_number)

SELECT product_number, product_name, vendor_name, wholesale_price  FROM product
NATURAL JOIN q
NATURAL JOIN product_vendor
NATURAL JOIN vendor
WHERE wholesale_price = lowest_price
ORDER BY product_number
```

```python
## Problem 11: List all unique pairs of orders that have the same date, customer, and employee. 
```

```python
%%sql

## o1.order_number < o2.order_number ensures unique pairs

SELECT * FROM `order` AS o1 
JOIN `order` AS o2 
ON o1.order_date = o2.order_date
AND o1.customer_id = o2.customer_id
AND o1.employee_id = o2.employee_id
WHERE o1.order_number < o2.order_number
```

```python
## Problem 12: Show all the triplets of products that appear together on at least 100 unique orders.
```

```python
%%sql

SELECT 
    i1.product_number as p1,
    i2.product_number as p2,
    i3.product_number as p3, count(*) as n
    
    FROM
    (order__item as i1 JOIN order__item as i2 USING (order_number)) 
    JOIN order__item as i3 USING (order_number) 
    WHERE i1.product_number < i2.product_number
        AND i2.product_number < i3.product_number
    GROUP BY p1, p2, p3
    HAVING n > 100
```

```python
## Problem 13: Find the most common list of products on an order.
```

```python
%%sql

SELECT 
    COUNT(order_number) AS count, product_group 
FROM (
    SELECT order_number, GROUP_CONCAT(product_number ORDER BY product_number) AS product_group
    FROM order__item
    GROUP BY order_number) AS subquery
GROUP BY product_group
ORDER BY count DESC
LIMIT 1;
```

```python
## Problem 14: Show all guests who have made reservations for any nights when Samantha Smith made reservations. 
```

```python
%%sql

WITH q AS(
SELECT date, guest_id
FROM reservation 
JOIN guest USING(guest_id)
WHERE guest_name = 'Samantha Smith')

SELECT guest_name, date FROM reservation
JOIN guest USING(guest_id) WHERE guest_id NOT IN(
SELECT guest_id FROM q) AND 
date IN (SELECT date FROM q)
GROUP BY guest_id
ORDER BY date
```

```python
## Problem 15: Find the pair of guests who have made reservations for the same night the most. 
```

```python
%%sql

SELECT 
    r1.guest_id as g1,
    r2.guest_id as g2,
    count(*) as reservations
    
    FROM
    (reservation as r1 JOIN reservation as r2 USING (date))  
    WHERE r1.guest_id < r2.guest_id
    GROUP BY g1, g2
    ORDER BY reservations DESC
    LIMIT 1;
    
```
